```python
# goal is to add head coach column to nflscrapR data

# i'll assemble a dataframe that has the following cols:

# [season     week      team       coach]     

# there is definitely a good way to do this with a dataframe pivot but i'll do the slower simple dumber way with 3 for loops
coach_df = []
i = 0
for season in seasons:
    sea_coaches = coaches.loc[coaches.Season==season]
    for index, row in coaches.iterrows():
        season = row['Season']
        begin = row['Begin']
        end = row['End'] + 1
        coach = row['Coach']
        team = row['Tm']
        weeks = list(range(begin, end))
        for week in weeks:
            coach_df.append([season, week, team, coach])
            
coach_df = pd.DataFrame(coach_df, columns=['season', 'week', 'posteam', 'head_coach'])

coach_df.head()
```

```python
# weakly informative prior

import scipy.stats as ss

def plot_beta(x_range, a, b, mu=0, sigma=1, cdf=False, **kwargs):
    '''
    Plots the f distribution function for a given x range, a and b
    If mu and sigma are not provided, standard beta is plotted
    If cdf=True cumulative distribution is plotted
    Passes any keyword arguments to matplotlib plot function
    '''
    x = x_range
    if cdf:
        y = ss.beta.cdf(x, a, b, mu, sigma)
    else:
        y = ss.beta.pdf(x, a, b, mu, sigma)
    plt.plot(x, y, **kwargs)
    
x = np.linspace(0, 1, 2000)

plot_beta(x, 3, 3, 0, 1, color='red', lw=2, ls='-', alpha=0.5, label='pdf')
plot_beta(x, 3, 3, 0, 1, cdf=True, color='blue', lw=2, ls='-', alpha=0.5, label='cdf')
plt.legend();
```

Instead of ML models I'll use pymc3 to estimate the base rate.

```python
import pymc3 as pm, theano.tensor as tt

# total number of plays
n_plays = len(pbp)

# observed data

# covariates
X = mdf[['yards_for_td','ydstogo','down','half','half_seconds_remaining','score_differential']].values

plays = mdf.play_type.values

with pm.Model() as model:
    
    # basic linear regression
    # slope and intercept priors (very wide)
    m = pm.Normal('m', 0, sigma=10, shape=X.shape[1])
    
    b = pm.Normal('b', 0, sigma=10)
    
    # X is a matrix of our covariates. For each play we will give the model these individual covariates (features)
    # features: situation + coaching + quarterback
    # this is linear regression
    mu = b + tt.dot(X,m)
    
    # for binomial data (pass vs run), we need a link function. 
    theta = pm.Deterministic('theta', pm.math.invlogit(mu))
    
    # final estimation of probability that a team will pass
    binom = pm.Binomial('binom', n=n_plays, p=theta, observed=plays)

    
with model:
    trace = pm.sample(3000, tune=1000, cores=4)



```
